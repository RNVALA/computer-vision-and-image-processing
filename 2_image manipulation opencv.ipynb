### Fliping the image
- OpenCv has several ways to flip an image, we cann use the flip() function
- The parameter is the flipCode-flipcode is the value indicating what kind of flip we would like to perform
     - flipcode=0 flip vertically
     - flipcode>0 flip horizontally around y-axis positive value
     - flipcode<0 flip vertically and horizontally,flipping around both axes negative vallue
     


```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
image=cv2.imread('cat.png')
for flipcode in [0,1,-1]:
    im_flip =  cv2.flip(image,flipcode )
    plt.imshow(cv2.cvtColor(im_flip,cv2.COLOR_BGR2RGB))
    plt.title("flipcode: "+str(flipcode))
    plt.show()
```

```python
# we can also use the rotate() function
im_flip=cv2.rotate(image,0)
plt.imshow(cv2.cvtColor(im_flip,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
flip = {"ROTATE_90_CLOCKWISE":cv2.ROTATE_90_CLOCKWISE,
        "ROTATE_90_COUNTERCLOCKWISE":cv2.ROTATE_90_COUNTERCLOCKWISE,
        "ROTATE_180":cv2.ROTATE_180
       }
```

```python
for key,value in flip.items():
    plt.subplot(1,2,1)
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.title('original')
    plt.subplot(1,2,2)
    plt.imshow(cv2.cvtColor(cv2.rotate(image,value),cv2.COLOR_BGR2RGB))
    plt.title(key)
    plt.show()
```

## Croping an Image
- Cropping is "cutting out" the part of the image and throwing out the rest; we can crop using arrays. Let start with a vertical crop; the variable upper is the first row that we would like to include in the image, the variable lower is the last row we would like to include. We then use slicing to obtain the new imag

```python
upper=150
lower=400
crop_top=image[upper:lower,:,:]
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(crop_top,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#horizontal crop
left=150
right=400
crop_horizontal=crop_top[:,left:right,:]
plt.figure(figsize=(5,5))
plt.imshow(cv2.cvtColor(crop_horizontal,cv2.COLOR_BGR2RGB))
plt.show()
```

## Changing Specific Image Pixels

```python
array_sq=np.array(image)
array_sq[upper:lower,left:right,:]=0
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.title("orignal")
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(array_sq,cv2.COLOR_BGR2RGB))
plt.title("Altered Image")
plt.show()
```

- we can also create shapes and OpenCv,we can use the method rectangle. The parameter pt1 is the top-left cordinate of the rectangle:(left,top) or (x0,y0),pt2 is the botton right cordinate(right,lower) or(x1,y1). The parameter color is a tuple representing the intensity of each chanel (blue,green,red). Finally we have the line thickness.

```python
start_point,end_point=(left,upper),(right,lower)
image_draw=np.array(image)
cv2.rectangle(image_draw,pt1=start_point,pt2=end_point,color=(0,255,0),thickness=3)
plt.figure(figsize=(5,5))
plt.imshow(cv2.cvtColor(image_draw,cv2.COLOR_BGR2RGB))
plt.show()
```

- we can overlay text on an image using the function putText with the parameter values:
   - img:image array
   - text: Text string to be overlayed
   - org: Bottom-left corner of the text string in the image
   - fontFace:type of font
   - fontScale:fontscale
   - color: Textcolor
   - thickness:Thickness of the lines used to draw a text
   - lineType:line type
   

```python
image_draw=cv2.putText(img=image,text='cat',org=(10,500),color=(255,255,255),fontFace=3,fontScale=10,thickness=2)
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.show()
```
